### Sistemazione dei valori contenuti nelle colonne 'world_rank' delle tabelle 'shanghai_ranking' e 'times_ranking'

Alcuni valori della colonna 'world_rank' della tabella 'shanghai_ranking' sono compresi in un intervallo. Per questo motivo, il tipo di dato contenuto nella colonna in questione è di tipo object, come si può vedere di seguito:

```python
shanghai_ranking[shanghai_ranking['world_rank'].str.contains('-')==True]['world_rank'].head(3)
```

Per tale ragione definisco la funzione 'string_to_number' che converte i ranking non espressi in range in valori interi. Nel caso in cui il ranking è espresso attraverso un intervallo di valori, la funzione individua gli estremi dell'intervallo, converte tali estremi in numeri interi calcolando successivamente il valore centrale dell'intervallo.


```python
def string_to_number(column_value):
    
    if len(column_value) <= 3:
        return int(column_value)
    
    if len(column_value) == 7:
        pattern = re.compile("(\d+)\-(\d+)")
        occurrence = pattern.search(column_value)
        
        if occurrence:
            lower = int(occurrence.group(1))
            upper = int(occurrence.group(2))
            central_value = (upper + lower) / 2
            return int(central_value)
```

Adesso applico la funzione 'string_to_number' alla colonna 'world_rank' di 'shanghai_ranking', assegnando la Series risultante dall'applicazione ad una nuova colonna della tabella in questione, denominata 'world_number_rank'.

```python
shanghai_ranking['world_number_rank'] = shanghai_ranking['world_rank'].apply(string_to_number)
```

### Normalizzazione dei nomi delle università per le tre tabelle di ranking relative ai valori più recenti                               per ogni università attraverso la definizione di una funzione.

#### Perchè questa funzione

Come abbiamo visto, i nomi delle università contenuti nella colonna university_name di ciascuna delle tre tabelle di ranking relative ai dati più recenti possono essere stati scritti in maniera differente. Per questo motivo, utilizzo la colonna 'school_name' del dataset 'school_and_country_table' assegnato alla variabile 'university_and_country' per normalizzare (i.e. rendere identici) i nomi delle università per tutte le tre suddette tabelle di ranking.

Dunque,l'idea è quella di creare per ciascuna delle tre tabelle di ranking una colonna che contenga i nomi delle università contenuti nella colonna 'university_name_normalized' della tabella university_and_country, ovviamente corrispondenti, riga per riga,all'università indicata nella colonna 'university_name' della tabella di ranking considerata. In questo modo le tre tabelle conterrebbero nella rispettiva colonna creata i nomi delle università scritti allo stesso modo.
Per questo motivo, tale colonna in tutte le tre tabelle verrà nominata 'university_name_normalized'.

Dunque, procedo ad importare la tabella 'school_and_country_table', rinominando opportunamente la colonna relativa ai nomi delle università che verrà utilizata per la normalizzazione, secondo quanto sopra affermato.

#### Funzionamento di base della funzione to_normalize

Per questa ragione, ho ideato la funzione to_normalize, che applico alla colonna 'university_name' di ciascuna tabella di ranking di dati più recenti, tabella per tabella.

Di seguito viene riportata la spiegazione del funzionamento di base della funzione to_normalize e, nella cella successiva, il relativo codice python.

Questa funzione prende in input, ciascuno per volta, i nomi delle università contenuti nella suddetta colonna della tabella di ranking. Successivamente, tale funzione controlla se il nome dell'università in input coincide con un nome presente nella colonna university_name_normalized di university_and_country, per ogni nome delle università contenute in quest'ultima colonna.

Successivamente, se c'è uguaglianza, la funzione restituisce il nome dell'università in input poichè di fatto i nomi delle unviersità delle due colonne confrontate sono scritti allo stesso modo. Successivamente a tale uguaglianza la funzione prende in input il valore successivo della colonna della tabella di ranking.

Se non ci fosse uguaglianza tale funzione definisce una Regular Expression (REGEX). Il pattern dellla REGEX è proprio il nome dell'università in input. Con un ciclo for, definisco volta per volta la stringa di occorrenza che è il nome dell'università   contenuta nella colonna university_name_normalized. 

Si noti che il pattern è al di fuori del ciclo for. 
Infatti,per ogni nome di quest'ultima colonna, la funzione individua se il nome dell'università in input è contenuto per intero nel nome  dell'università della colonna university_name_normalized considerato dal ciclo for. In caso di occorrenza, la funzione restituisce il nome dell'università contenuto in university_name_normalized, appunto per avere normalizzazione della colonna di input.

Se il nome dell'università in input non è contenuto nel nome dell'università della colonna university_name_normalized, la funzione esegue la regex al contrario solo per lo stesso nome dell'università di university_name_normalized considerato con il ciclo for. In caso di occorrenza, la funzione restituisce il nome dell'università contenuto in university_name_normalized, appunto per avere normalizzazione della colonna di input. In caso di mancata occorrenza, il ciclo for considera il nome successivo dell'università presente nella colonna university_name_normalized e la funzione applica di nuovo la regex e regex inversa, finchè non c'è occorrenza oppure finchè non termina il ciclo for, che rappresenta una mancanza di occorrenza in ogni caso.

Se termina il ciclo for a causa di una mancata occorrenza, la funzione restituisce il dato None, che indica l'assenza di match in ogni caso individuato dalla funzione.

```python
def basic_to_normalize_function (column_value):
    
    for university in university_and_country['university_name_normalized']:
        if column_value == university:
            return column_value
        
    pattern = re.compile(column_value)
    for university in university_and_country['university_name_normalized']:
        occurrence = pattern.search(university)
        if occurrence:
            return university
        else:
            reverse_pattern = re.compile(university)
            reverse_occurrence =reverse_pattern.search(column_value)
            if reverse_occurrence:
                return university
            
```

#### Valori duplicati come indicatori di modifica della funzione

Tuttavia la funzione potrebbe generare , come risultato finale dell'applicazione, valori doppi nella colonna risultante normalizzata,individuati tramite la funzione .duplicated().

Questo potrebbe essere dato  principalmente dai seguenti motivi:

1) la colonna university_name della tabella di ranking considerata oltre a contenere il nome di una data università contiene anche i nomi dei rispettivi campus, generando valori doppi che non consideriamo in quanto la tabella university_and_country considera solamente l'università nel suo complesso,senza prendere in considerazione campus e centri di ricerca della stessa università.

2) il nome di una data università nella colonna university_name è stata denominata negli anni in modo differente. Di fatto, tale università viene considerata come se fosse una diffente università, generando valori doppi, le cui righe corrispondenti vanno eliminate dalla tabella dei dati più recenti. Tali righe devono essere eliminate in quanto esse fanno riferimento ad un rank di un anno meno recente per la stessa unversità in questione.

3) il nome dell'università nella colonna university_name presenta caratteri di codifica differenti (ad esempio una parte del nome dell'università è posta a capo della cella) rispetto al nome corrispondente contenuto nella colonna university_name_normalized della tabella university_and_country, nonostante i due nomi siano scritti in modo identico.

4) la funzone genera errori nel match dei due nomi delle università in quanto i nomi sono molto simili ma diversi.


Per questo motivo ogni funzione applicata a ciascuna tabella presenterà una modifica particolare per evitare di incorrere in errore. tali modifiche le ho eseguite gestendo i valori doppi manualmente, uno per volta e cercando sul web se un dato nome corrispondesse ad un università diversa da quella matchata o se fosse un campus differente.

### Funzione che determina la massima differenza tra i ranking

La seguente funzione viene eseguita su ciascuna riga del dataframe shanghai_times_world_university, che dunque rappresentano, volta per volta, il valore di input della funzione medesima. In particolare, per ciascuna riga, la funzione individua il ranking di shanghai, times e world presenti nella suddetta tabella, assegnando ciascuno alla rispettiva variabile identificatrice del ranking medesimo. Successivamente, dati tali tre valori, la funzione cerca nell'insieme di tali valori numerici rispettivamente il  ranking massimo e il  ranking minimo,  in modo tale da restituire la massima differenza possibile tra i tre ranking in questione.

```python
def maximum_difference_between_rankings(row):
    
    shanghai_ranking = row['shanghai_ranking']
    times_ranking = row['times_ranking']
    world_ranking = row['world_ranking']
    
    maximum_ranking = max(shanghai_ranking,times_ranking,world_ranking)
    minimum_ranking = min(shanghai_ranking,times_ranking,world_ranking)
    
    return (maximum_ranking - minimum_ranking)
```

Successivamente, applico la funzione 'maximum_difference_between_rankings' al dataframe shangai_times_world_university e creo una nuova colonna del dataframe medesimo, nominata 'highest_ranking_difference', al quale assegno la Series di valori restituita dalla funzione in questione.

```python
shanghai_times_world_university ['highest_ranking_difference'] = shanghai_times_world_university.apply(maximum_difference_between_rankings,axis=1)
shanghai_times_world_university 
```

Adesso definisco la funzione 'number_of_female' che mi estrae la frazione di studentesse, sul totale degli studenti di ciascuna università, dai valori della colonna 'female_male_ratio' della tabella in questione. La funzione permette di moltiplicare tale frazione per il numero totale degli studenti di ciascuna università in questione, ottenendo in questa maniera il numero totale di studentesse.

```python
def number_of_female(row):
    
    if (row['first_condition'] == True) or (row['second_condition'] == True):
        return None
        
    pattern = re.compile('^([0-9]+)')
    occurrence = pattern.search(row['female_male_ratio'])
    
    if occurrence:
        fraction_female = int(occurrence.group(1)) / 100
        total_students = row['num_students']
        number_female = round(fraction_female * total_students,0)
        
        return number_female
```

```python
times_most_recent_data['number_of_female_students']=times_most_recent_data.apply(number_of_female,axis=1)
```

Di seguito definisco la funzione 'number_of_male' che calcola il numero di studenti maschi per ciascuna università, secondo la  stessa metodologia introdotta per la funzione 'number_of_female'.

```python
def number_of_male(row):
    
    if (row['first_condition'] == True) or (row['second_condition'] == True):
        return None
    
    pattern = re.compile('([0-9]+)$')
    occurrence = pattern.search(row['female_male_ratio'])
    
    if occurrence:
        fraction_male = int(occurrence.group(1))/100
        total_students = row['num_students']
        number_male = round(fraction_male * total_students,0)
        
        return number_male
```

Adesso applico la funzione alla colonna 'female_male_ratio' e la Series risultante dall'applicazione della funzione la assegno ad una nuova colonna della tabella 'times_most_recent_data', denominata 'number_of_male_students'.

Lo scopo della definizione di tale funzione è impedire la produzione di valori 'inf' quando il denominatore del rapporto è pari a zero.

```python
def to_compute_female_male_ratio(row):
    if  (row['number_of_male_students'] == None) or (row['number_of_male_students'] == 0):
        return None
    else:
        return round(row['number_of_female_students'] / row['number_of_male_students'] *100,1)
```
